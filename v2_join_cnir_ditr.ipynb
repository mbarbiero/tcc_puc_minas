```python
##
#   Faz o join entre o CNIR e a DITR2020 e
#   e cria o conjunto de dados falsos
##
import numpy as np
import pandas as pd
import os
```

```python
# Conjunto de cnirs vinculados. Esses registros serão considerados como Verdadeiro.
cnirs_v = pd.read_pickle('..\\DADOS\\cnirs_trein.pickle') 
# Conjunto de toas as DITRs do ano base 2020
ditr2020 = pd.read_pickle('..\\DADOS\\ditr2020.pickle')
```

```python
# cnirs_ditr2020_v é o conjunto de dados que foram classificados como corretos (class = True)
cnirs_ditr2020_v = cnirs_v.merge(ditr2020, left_on='cnir.nirf', right_on='ditr2020.nirf', how='inner')
cnirs_ditr2020_v['class'] = True
cnirs_ditr2020_v.to_pickle('..//DADOS//v2_cnirs_DITR2020_v.pickle')
cnirs_ditr2020_v
```

```python
# cnirs_ditr2020_f é o conjunto de dados que foram classificados como incorretos (class = False)
cnirs_f = cnirs_v.copy(deep=True)
# o método join une os conjuntos a partir do index e produz um resultado quase aleatório
cnirs_ditr2020_f = cnirs_f.join(ditr2020, how='inner')
cnirs_ditr2020_f['class'] = False
cnirs_ditr2020_f.to_pickle('..//DADOS//v2_cnirs_DITR2020_f.pickle')
cnirs_ditr2020_f
```

```python
## Criando um subconjunto de aprendizado
cj_aprend = cnirs_ditr2020_v.append(cnirs_ditr2020_f)
print(cj_aprend.shape)
```

```python
# Criando colunas igualdade (ig.xxxxx)
cj_aprend['ig.ni'] = cj_aprend['cnir.ni'] == cj_aprend['ditr2020.ni']
cj_aprend['ig.codigoIncra'] = cj_aprend['cnir.codigoIncra'] == cj_aprend['ditr2020.codigoIncra']
cj_aprend['ig.municipio'] = cj_aprend['cnir.municipio'] == cj_aprend['ditr2020.municipio']

# Criando coluna de variação da área entre CNIR e DITR2019
#cj_aprend[['cnir.areaTotal', 'ditr2019.areatotal']] = cj_aprend[['cnir.areaTotal', 'ditr2019.areatotal']].apply(pd.to_numeric)
cj_aprend['dif'] = np.where(cj_aprend['cnir.areaTotal'] > cj_aprend['ditr2020.areaTotal'], 
             (cj_aprend['cnir.areaTotal'] - cj_aprend['ditr2020.areaTotal']) / cj_aprend['cnir.areaTotal'], 
             (cj_aprend['ditr2020.areaTotal'] - cj_aprend['cnir.areaTotal']) / cj_aprend['ditr2020.areaTotal'])  
#reduzindo a quantidade  de valores para simplificar o aprendizado
cj_aprend['dif'] = cj_aprend['dif'].apply(lambda x:int(x * 100))

cj_aprend
```

```python
# Eliminação das colunas desnecessárias na aprendizagem
del cj_aprend['cnir.nirf']
del cj_aprend['cnir.municipio']
del cj_aprend['cnir.areaTotal']
del cj_aprend['cnir.ni']
del cj_aprend['cnir.codigoIncra']

del cj_aprend['ditr2020.ni']
del cj_aprend['ditr2020.areaTotal']
del cj_aprend['ditr2020.municipio']

cj_aprend.to_pickle('..//DADOS//v2_cj_aprend.pickle')

```
