```python
##
#   Junta as bases do CNIR e DITR por meio do nirf
##
import numpy as np
import pandas as pd
import os
```

```python
cnirs = pd.read_pickle('..\\DADOS\\cnirs_sem_vinculacao.pkl')
ditr2020 = pd.read_pickle('..\\DADOS\\ditr2020.pickle')
cnirs
```

```python
cnirs_ditr2020_por_ni = cnirs.merge(ditr2020, left_on='cnir.ni', right_on='ditr2020.ni', how='inner')
cnirs_ditr2020_por_ni.to_pickle('..//DADOS//v2_cnirs_ditr2020_por_ni.pkl')
cnirs_ditr2020_por_ni
```

```python
## Criando dados para aplicação do modelo
cj_aplic = pd.read_pickle('..//DADOS//v2_cnirs_ditr2020_por_ni.pkl')

# Criando colunas igualdade (ig.xxxxx)
cj_aplic['ig.ni'] = cj_aplic['cnir.ni'] == cj_aplic['ditr2020.ni']
cj_aplic['ig.codigoIncra'] = cj_aplic['cnir.codigoIncra'] == cj_aplic['ditr2020.codigoIncra']
cj_aplic['ig.municipio'] = cj_aplic['cnir.municipio'] == cj_aplic['ditr2020.municipio']

# Criando coluna de variação da área entre CNIR e DITR2019
#cj_aprend[['cnir.areaTotal', 'ditr2019.areatotal']] = cj_aprend[['cnir.areaTotal', 'ditr2019.areatotal']].apply(pd.to_numeric)
cj_aplic['dif'] = np.where(cj_aplic['cnir.areaTotal'] > cj_aplic['ditr2020.areaTotal'], 
             (cj_aplic['cnir.areaTotal'] - cj_aplic['ditr2020.areaTotal']) / cj_aplic['cnir.areaTotal'], 
             (cj_aplic['ditr2020.areaTotal'] - cj_aplic['cnir.areaTotal']) / cj_aplic['ditr2020.areaTotal'])  
#reduzindo a quantidade  de valores para simplificar o aprendizado
cj_aplic['dif'] = cj_aplic['dif'].apply(lambda x:int(x * 100))

cj_aplic
```

```python
# Eliminação das colunas desnecessárias
del cj_aplic['cnir.nirf']
del cj_aplic['cnir.municipio']
del cj_aplic['cnir.areaTotal']
del cj_aplic['cnir.ni']
#del cj_aplic['cnir.codigoIncra']

del cj_aplic['ditr2020.ni']
del cj_aplic['ditr2020.areaTotal']
del cj_aplic['ditr2020.municipio']

cj_aplic.to_pickle('..//DADOS//v2_cj_aplic.pkl')

```
