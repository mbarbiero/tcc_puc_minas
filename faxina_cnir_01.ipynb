```python
##
#   Padroniza o formato dos valores do CNIR
##
import numpy
import pandas as pd
import os

arquivo = 'd:\\temp\\cnir\\cnirs.csv'
cnirs = pd.read_csv(arquivo, dtype={
                     'cnir.nirf': str,
                     'cnir.codigoIncra': str,
                     'cnir.ni': str,
                     'cnir.municipio': str,
                     'cnir.areaTotal': float
                 })

# Removendo a coluna 'index'
cnirs.drop(cnirs.columns[[0]], axis=1, inplace=True)

cnirs
```

```python
# Renomeando colunas 
cnirs.rename(columns={'codigoCnir': 'cnir.codigoCnir'}, inplace = True)
cnirs.to_csv(r'd:\\temp\\cnir\\cnirs.csv')
```

```python
# Selecionando registros para criar base de aprendizado
cnirs_trein = cnirs[(cnirs['cnir.nirf'].str.len() == 8)]    # tamanho do campo com apenas um nirf vinculado
cnirs_trein
```

```python
# Eliminando nirfs repetidos (1 x n)
nirfs_repetidos = cnirs_trein[cnirs_trein.duplicated(['cnir.nirf'], keep=False)]
cnirs_trein = cnirs_trein.drop(nirfs_repetidos.index)

cnirs_trein.to_pickle('..\\DADOS\\cnirs_trein.pickle')
cnirs_trein
```
