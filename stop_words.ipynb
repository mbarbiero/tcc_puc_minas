```python
##
#   Consolida os arquivos de dados municipais baixados da base CNIR
#   Seleciona os campos de denominação e localização para futura creiação de stop words
##
import pandas as pd
from nltk.tokenize import word_tokenize

frases = pd.read_csv(r'd:\\temp\\cnir\\identificacao_imoveis.csv')
frases 
```

```python
##
#   Criação da lista de stopwords da denominação do imóvel com nltk
##
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from string import punctuation
import csv

denominacoes = frases['denominacao']
denominacoes.dropna(inplace=True)
palavras = word_tokenize(denominacoes.to_string(index=False).lower())

# elimina as stropwords da lingua portuguesa e a pontuação
stopwords = set(stopwords.words('portuguese') + list(punctuation))
palavras_sem_stopwords = [palavra for palavra in palavras if palavra not in stopwords]

# salva a lista de palavras 
arq_palavras = pd.DataFrame(palavras_sem_stopwords)
arq_palavras.to_csv('..\DADOS\palavras_sem_stopwords.csv', index=False)

# calcula a frequencia de palavras 
frequencia = FreqDist(palavras_sem_stopwords)
frequencia.most_common(50)
```
